```python
import numpy as np
import matplotlib.pyplot as plt
from method.Optimizer import Optimizer
```

```python
def objective(x: list) -> float:
    """
    Define objective function
    :param x: Coordinates of points
    :return: Loss value of objective function
    """
    return 0.0001965039799 * x[0] ** 6 - 0.0030680488103 * x[0] ** 5 + 0.0139137087473 * x[0] ** 4 - 0.0164183206347 * x[0] ** 3 - 0.1141273130213 * x[0] ** 2 - 0.3051376566158 * x[0] + 8.0462421383821

def derivative(x: list) -> np.ndarray:
    """
    Define derivative of objective function
    :param x: Coordinates of points
    :return: The partial derivatives of the objective function at the point with respect to each parameter
    """
    return np.asarray([
        -0.3051376566158 - 0.2282546260426 * x[0] - 0.0492549619041 * x[0] ** 2 + 0.0556548349892 * x[0] ** 3 - 0.015340244052 * x[0] ** 4 + 0.00117902 * x[0] ** 5])

# Define learning rate and its decay factor
alpha, gamma = 0.3, 1
# Setting the hyperparameter betas
beta_1, beta_2, beta_3 = 0.9, 0.5, 0.5
# Initialise the optimiser
optimizer = Optimizer(objective=objective, derivative=derivative, steps=50, start_point=[-4.5])
# Training by SGD
track_list_sgd, loss_val_list_sgd = optimizer.sgd(alpha=alpha, gamma=gamma)
track_list_sgd = np.asarray(track_list_sgd)
# Training by momentum
track_list_momentum, loss_val_list_momentum = optimizer.momentum(alpha=alpha, gamma=gamma, beta_1=beta_1)
track_list_momentum = np.asarray(track_list_momentum)
# Training by RMSprop
track_list_rmsprop, loss_val_list_rmsprop = optimizer.rmsprop(alpha=alpha, gamma=gamma, beta_2=beta_2)
track_list_rmsprop = np.asarray(track_list_rmsprop)
# Training by adam
track_list_adam, loss_val_list_adam = optimizer.adam(alpha=alpha, gamma=gamma, beta_1=beta_1, beta_2=beta_2)
track_list_adam = np.asarray(track_list_adam)
# Training by signum
track_list_signum, loss_val_list_signum = optimizer.signum(alpha=alpha, gamma=gamma, beta_1=beta_1)
track_list_signum = np.asarray(track_list_signum)
# Training by lion
track_list_lion, loss_val_list_lion = optimizer.lion(alpha=alpha, gamma=gamma, beta_1=beta_1, beta_3=beta_3)
track_list_lion = np.asarray(track_list_lion)
```

```python
# Generate points on the objective function uniformly within bounds
x_coord = np.linspace(-5, 10, 400)
y_coord = [objective([x]) for x in x_coord]
# Creating a Canvas
fig = plt.figure(figsize=(14, 15))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)
# Plot the decline curve of SGD
ax1.plot(x_coord, y_coord, label='objective function')
ax1.plot(track_list_sgd[:, 0], loss_val_list_sgd, '.-', label='SGD')
ax1.legend()
ax1.grid(True)
# Plot the decline curve of momentum
ax2.plot(x_coord, y_coord, label='objective function')
ax2.plot(track_list_momentum[:, 0], loss_val_list_momentum, '.-', label='momentum')
ax2.legend()
ax2.grid(True)
# Plot the decline curve of RMSprop
ax3.plot(x_coord, y_coord, label='objective function')
ax3.plot(track_list_rmsprop[:, 0], loss_val_list_rmsprop, '.-', label='RMSprop')
ax3.legend()
ax3.grid(True)
# Plot the decline curve of adam
ax4.plot(x_coord, y_coord, label='objective function')
ax4.plot(track_list_adam[:, 0], loss_val_list_adam, '.-', label='adam')
ax4.legend()
ax4.grid(True)
# Plot the decline curve of adam
ax5.plot(x_coord, y_coord, label='objective function')
ax5.plot(track_list_signum[:, 0], loss_val_list_signum, '.-', label='signum')
ax5.legend()
ax5.grid(True)
# Plot the decline curve of adam
ax6.plot(x_coord, y_coord, label='objective function')
ax6.plot(track_list_lion[:, 0], loss_val_list_lion, '.-', label='lion')
ax6.legend()
ax6.grid(True)

plt.show()
```
